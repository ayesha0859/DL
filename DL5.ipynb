```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences

vocabulary_size = 10000
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=vocabulary_size)

print(f"Training sequences: {len(x_train)}")
print(f"Test sequences: {len(x_test)}")
print(f"Example review (first 10 words): {x_train[0][:10]}")
print(f"Example label: {y_train[0]}")
```

```python
max_sequence_length = 250 # Each review will be padded/truncated to this length

x_train_padded = pad_sequences(x_train, maxlen=max_sequence_length)
x_test_padded = pad_sequences(x_test, maxlen=max_sequence_length)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

embedding_dim = 128

model = Sequential([
    Embedding(vocabulary_size, embedding_dim),
    SimpleRNN(128),
    Dense(1, activation='sigmoid')
])
model.build(input_shape=(None, max_sequence_length))
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
epochs = 10
batch_size = 32

history = model.fit(x_train_padded, y_train,
                    epochs=epochs,
                    batch_size=batch_size,
                    validation_split=0.2) # Use 20% of training data for validation
```

```python
loss, accuracy = model.evaluate(x_test_padded, y_test)

print(f"\nTest Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```

```python
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences
import string

try:
    word_index
except NameError:
    from tensorflow.keras.datasets import imdb
    word_index = imdb.get_word_index()

def preprocess_new_review(review_text, word_index, max_len):
    review_text = review_text.lower().translate(str.maketrans('', '', string.punctuation))

    words = review_text.split()
    encoded_review = [word_index.get(word, 2) + 3 for word in words]
    padded_review = pad_sequences([encoded_review], maxlen=max_len)
    return padded_review[0]

user_review_text = "This movie was absolutely fantastic! I loved every minute of it and the acting was superb."

if user_review_text:
    preprocessed_input = preprocess_new_review(user_review_text, word_index, max_sequence_length)

    input_for_model = np.expand_dims(preprocessed_input, axis=0)

    prediction = model.predict(input_for_model)[0][0]

    sentiment = "Positive" if prediction >= 0.5 else "Negative"

    print(f"\nReview: '{user_review_text}'")
    print(f"Prediction probability: {prediction:.4f}")
    print(f"Predicted sentiment: {sentiment}")
else:
    print("Please enter a movie review to get a prediction.")
```
