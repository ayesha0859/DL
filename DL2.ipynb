### 2. Keras (often integrated with TensorFlow)

```python
from tensorflow import keras
from tensorflow.keras import layers

# build model
model = keras.Sequential([
    layers.Dense(4, activation='relu', input_shape=(2,)),
    layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy')

model.summary()
```

### 4. JAX

```python
import jax
import jax.numpy as jnp

# Create a JAX array
x = jnp.array([1.0, 2.0, 3.0])
print(x)

# Apply a JAX transformation (e.g., JIT compilation)
@jax.jit
def sum_and_square(x):
    return jnp.sum(x ** 2)

print(sum_and_square(x))
```

### 5. Hugging Face Transformers

```python
# Install transformers library if not already installed
# !pip install transformers
```

```python
from transformers import pipeline

# Load a sentiment analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis")

# Analyze text
result = sentiment_analyzer("I love using deep learning libraries!")
print("Hugging face output:", result)

result = sentiment_analyzer("This is a terrible movie.")
print(result)
```
