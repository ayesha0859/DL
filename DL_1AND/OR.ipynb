```python
def water_needed_and(soil_low, temp_high):
    w1 = 1
    w2 = 1
    theta = 2

    net_input = (w1 * soil_low) + (w2 * temp_high)

    return 1 if net_input >= theta else 0


print("| SoilLow | TempHigh => WaterNeeded")
for x1 in [0, 1]:
    for x2 in [0, 1]:
        print(f"|   {x1}     |    {x2}     =>       {water_needed_and(x1, x2)}")
```

```python
def water_needed_or(soil_low, temp_high):
    w1 = 1
    w2 = 1
    theta = 1

    net_input = (w1 * soil_low) + (w2 * temp_high)

    return 1 if net_input >= theta else 0


print("\n| SoilLow | TempHigh => WaterNeeded")
for x1 in [0, 1]:
    for x2 in [0, 1]:
        print(f"|   {x1}     |    {x2}     =>       {water_needed_or(x1, x2)}")
```

```python
import numpy as np
```

```python
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])

weights = np.array([0.0, 0.0])
bias = 0.0
learning_rate = 0.1
epochs = 5
```

```python
print("Initial Weights:", weights)
print("Initial Bias:", bias)
print("-" * 60)

#training
for epoch in range(epochs):
    print(f"\nEpoch {epoch + 1}")
    print("-" * 60)

    for i in range(len(X)):

        net = np.dot(X[i], weights) + bias

        #activation
        prediction = 1 if net >= 0 else 0

        #error
        error = y[i] - prediction

        #update weights and bias
        weights = weights + learning_rate * error * X[i]
        bias = bias + learning_rate * error

        print(f"Input: {X[i]}")
        print(f"Target Output: {y[i]}")
        print(f"Net Input: {net:.2f}")
        print(f"Predicted Output: {prediction}")
        print(f"Error: {error}")
        print(f"Updated Weights: {weights}")
        print(f"Updated Bias: {bias}")
        print("-" * 60)

print("\nFinal Weights:", weights)
print("Final Bias:", bias)
```
