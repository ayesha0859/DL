```python
!pip install tensorflow

import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, GRU, Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
from tensorflow.keras.datasets import imdb
vocab_size = 5000
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=vocab_size)

print("Training samples:", len(X_train))
print("Testing samples:", len(X_test))
```

```python
max_len = 200

X_train = pad_sequences(X_train, maxlen=max_len)
X_test = pad_sequences(X_test, maxlen=max_len)
```

```python
model_lstm = Sequential([
    Embedding(input_dim=vocab_size, output_dim=64),
    LSTM(64),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

model_lstm.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
model_lstm.build(input_shape=(None, max_len))
model_lstm.summary()
```

```python
history_lstm = model_lstm.fit(
    X_train, y_train,
    epochs=5,
    batch_size=64,
    validation_data=(X_test, y_test)
)
```

```python
loss, acc = model_lstm.evaluate(X_test, y_test)
print("LSTM Accuracy:", acc)
```

```python
model_gru = Sequential([
    Embedding(input_dim=vocab_size, output_dim=64),
    GRU(64),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

model_gru.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
model_gru.build(input_shape=(None, max_len))
model_gru.summary()
```

```python
history_gru = model_gru.fit(
    X_train, y_train,
    epochs=5,
    batch_size=64,
    validation_data=(X_test, y_test)
)
```

```python
loss, acc = model_gru.evaluate(X_test, y_test)
print("GRU Accuracy:", acc)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

word_index = imdb.get_word_index()

reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])

def decode_review(text):
    return " ".join([reverse_word_index.get(i - 3, '?') for i in text])

tokenizer = Tokenizer(num_words=5000)

decoded_reviews = [decode_review(x) for x in X_train[:2000]]
tokenizer.fit_on_texts(decoded_reviews)

def predict_text(model, text):
    seq = tokenizer.texts_to_sequences([text])
    padded = pad_sequences(seq, maxlen=max_len)
    pred = model.predict(padded)
    return "Positive" if pred > 0.5 else "Negative"

my_review = "This movie was really amazing and enjoyable"
my_review2 = "This movie was really bad and disgusting"

print(predict_text(model_lstm, my_review))
print(predict_text(model_gru, my_review2))
```
